Nível de escolaridade	Inscritos

Fundamental completo	451

Médio completo	627

Superior completo	292

Pós-graduação completa	95

Com base na tabela acima, crie uma Series com esses valores e calcule:


(a) O número total de inscritos

(b) Número inscritos que possuem, pelo menos, o superior completo.

(c) Suponha que a inscrição foi prorrogada e, com isso, foi obtido um número adicional de inscrições: 
68, 93, 22 e 0 inscritos a mais em cada uma das categorias de escolaridade apresentados na tabela acima

```python
import pandas as pd

inscritos = pd.Series([451,627,292,95])
print('a) Total inscritos: ', inscritos.sum())
print('b) Inscritos com pelo menos superior completo: ', inscritos[2:].sum())
novas_inscricoes = pd.Series([68,93,22,0])
print('c) Novo total inscrições:', (inscritos+novas_inscricoes).sum() )
```

Uma turma de cinco alunos foi submetida a uma avaliação e a nota de cada um dos alunos é apresentada na tabela abaixo:

Nome do Aluno	Nota

Wilfred	2

Abbie	7

Harry	5.5

Julia	10

Carrie	6.2

(a) Crie uma Series com os valores da tabela acima.

(b) Retorne o número de alunos que foram aprovados (considerando que a nota mínima para a aprovação é 7).

(c) Qual foi a nota média da turma.

(d) Obtenha o nome dos alunos que tiraram notas acima da média da turma.


```python
notas = pd.Series([2,7,5.5,10,6.2])

reprovados = notas[notas <7].count()
print('Nº alunos reprovados:',reprovados)
print('Média da turma:',notas.mean())
alunos = pd.Series(['Wilfred','Albie','Harry','Julia','Carrie'])
notas = notas.set_axis(alunos)
print('Alunos com notas maiores que a média da sala:',list(notas[notas> notas.mean()].index))
```

```python
df = pd.DataFrame([[72,180,31],[80,170,21],[60,175,16],[90,174,28],[100,185,23]],columns=['Peso(kg)','Altura(cm)','Idade'])
display(df['Altura(cm)'])
print('Média da idade: ', df['Idade'].mean())
print(f'Pessoas abaixo da altura média ({df["Altura(cm)"].mean()}cm)')
display(df[df['Altura(cm)'] < df['Altura(cm)'].mean()])
df['IMC'] = round(df['Peso(kg)']/((df['Altura(cm)']/100)**2),2)
display(df)
df = df.drop(axis='columns',labels='Idade')
display(df)
df.to_csv(path_or_buf='./imc.csv',sep=',', index=False)
```
