**Desafio**

Temos uma meta de R$ 1.5 Mi, dado as vendas do ano:

- Quando a meta foi batida?
  - Lembre-se do cumsum, e mascaras
- Qual foi o dia de menor venda? Qual o valor?
  - Lembre-se do `argmin`
- Qual foi o dia de maior venda? Qual o valor?
  - Lembre-se do `argmax`
- Qual a média de vendas?
- Qual o desvio padrão de vendas?

Dica:
- Para o menor/maior dia de venda lembre-se do `argmin` e `argmax`, funções que retornam o indice que ocorreu tal fenômeno

```python
import pandas as pd
import numpy as np

np.random.seed(42)
meta = 1.5*10**6
vendas = np.random.normal(50_000, 18000, 61)
serie_vendas = pd.Series(vendas).astype(int)
serie_vendas.plot()
```

```python
vendas_acumuladas = serie_vendas.cumsum()
dia_meta_batida = vendas_acumuladas[vendas_acumuladas >= meta].index.min()

#print('Meta foi atingida no dia:', np.where(vendas_acumuladas >= meta)[0][0])
print('Meta foi atingida no dia:', dia_meta_batida)
print('Menor dia de venda:', np.argmin(vendas), 'valor: ',vendas[np.argmin(vendas)])
print('Maior dia de venda:', np.argmax(vendas), 'valor: ',vendas[np.argmax(vendas)])
print('média:', vendas.mean())
print('std:', vendas.std())
```
